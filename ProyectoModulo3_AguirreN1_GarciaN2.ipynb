## Simulaciones y Visualización de resultados de simulación

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
from scipy.integrate import odeint
```

```python
g = 9.81
m1 = 1
m2 = 1 
l1 = 1
l2 = 1
t = np.linspace(0, 100, 2500)
```

```python
def f(x,t):
    theta1 = x[0]
    theta2 = x[1]
    w1 = x[2]
    w2 = x[3]
    
    dtheta1 = w1
    dtheta2 = w2
    dw1 = (-g*(2*m1 + m2)*np.sin(theta1) - m2*g*np.sin(theta1 - 2*theta2) - 2*np.sin(theta1 - theta2)*m2*(((dtheta2)**2)*l2 + ((dtheta1)**2)*l1*np.cos(theta1 - theta2)))/(l1*(2*m1 + m2 - m2*np.cos(2*theta1 - 2*theta2)))
    dw2 = (2*np.sin(theta1 - theta2)*(((dtheta1)**2)*l1*(m1 + m2) + g*(m1 + m2)*np.cos(theta1) + ((dtheta2)**2)*l2*m2*np.cos(theta1 - theta2)))/(l1*(2*m1 + m2 - m2*np.cos(2*theta1 - 2*theta2)))
    
    return [dtheta1, dtheta2, dw1, dw2]
```

## Primera condición inicial

```python
x0 = np.array([np.pi/2, np.pi/2, 0, 0])
```

```python
sol = odeint(f,x0,t)
sol=odeint(f,x0,t)
theta1=sol[:,0]
theta2=sol[:,1]
w1=sol[:,2]
w2=sol[:,3]
```

```python
plt.figure(figsize=(9,9))
plt.plot(t,theta1,label='$\\theta_1$')
plt.plot(t,theta2,label='$\\theta_2$')
plt.xlabel('tiempo')
plt.legend(loc='best')
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(9,9))
plt.plot(t,w1,label='$\\omega_1$')
plt.plot(t,w2,label='$\\omega_2$')
plt.xlabel('tiempo')
plt.legend(loc='best')
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(9,9))
plt.plot(theta1, theta2, label='Ángulos')
plt.xlabel('$\\theta_1$')
plt.ylabel('$\\theta_2$')
plt.legend(loc = 'best')
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(9,9))
plt.plot(w1, w2, label='Velocidades angulares')
plt.xlabel('$\\omega_1$')
plt.ylabel('$\\omega_2$')
plt.legend(loc = 'best')
plt.grid()
plt.show()
```

Esto último es la razón por la cual las gráficas no son tan "bonitas/uniformes", sino que son "caóticas". Esto quiere decir que es un sistema que se commporta drasticamente distinto con pequeñas variaciones inciales.
